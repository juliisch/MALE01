```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

# Daten erstellen
data = {
    'Einkommen': [3000, 4000, 2000, 1000, 3500, 1500, 5000],
    'Alter': [25, 35, 22, 21, 40, 28, 45],
    'Beruf': ['Angestellter', 'Manager', 'Student', 'Student', 'Angestellter', 'Arbeiter', 'Manager'],
    'Schulden': ['Nein', 'Nein', 'Ja', 'Ja', 'Nein', 'Ja', 'Nein'],
    'Wohnort': ['Stadt', 'Stadt', 'Dorf', 'Dorf', 'Stadt', 'Dorf', 'Stadt'],
    'Ausgaben': [1500, 2000, 1200, 800, 1600, 900, 2500],
    'Sparquote': [0.2, 0.3, 0.1, 0.05, 0.25, 0.1, 0.4],
    'Kreditwuerdig': ['Ja', 'Ja', 'Nein', 'Nein', 'Ja', 'Nein', 'Ja']
}

# DataFrame erstellen
df = pd.DataFrame(data)

# Label Encoding fuer kategoriale Variablen
le = LabelEncoder()
df['Beruf'] = le.fit_transform(df['Beruf'])
df['Schulden'] = le.fit_transform(df['Schulden'])
df['Wohnort'] = le.fit_transform(df['Wohnort'])
df['Kreditwuerdig'] = le.fit_transform(df['Kreditwuerdig'])

# Features und Zielvariable definieren
X = df[['Einkommen', 'Alter', 'Beruf', 'Schulden', 'Wohnort', 'Ausgaben', 'Sparquote']]
y = df['Kreditwuerdig']

# Trainings- und Testdaten aufteilen
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Decision Tree Classifier erstellen und trainieren
clf = DecisionTreeClassifier(criterion='entropy', max_depth=None, min_samples_split=2, random_state=42)
clf = clf.fit(X_train, y_train)

# Entscheidungsbaum visualisieren
plt.figure(figsize=(15,10))
tree.plot_tree(clf, feature_names=X.columns, class_names=['Nein', 'Ja'], filled=True)
plt.show()

# Vorhersage anzeigen
y_pred = clf.predict(X_test)
print("Vorhersagen:", y_pred)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Datensatz erstellen
data = {
    "Feature1": [5, 8, 2, 5],
    "Feature2": [2, 6, 4, 6],
    "Feature3": [9, 3, 3, 1],
    "Target":   ["A", "B", "B", "A"]
}


dataset = pd.DataFrame(data)

# Features und Zielvariable definieren
X = dataset[["Feature1", "Feature2", "Feature3"]]
y = dataset["Target"]

# Entscheidungsbaum-Modell erstellen und trainieren
decision_tree = DecisionTreeClassifier(criterion="gini", max_depth=2, random_state=42)
decision_tree.fit(X, y)

# Entscheidungsbaum visualisieren
plt.figure(figsize=(8, 6))
plot_tree(decision_tree, feature_names=X.columns, class_names=decision_tree.classes_, filled=True)
plt.title("Entscheidungsbaum")
plt.show()
```
